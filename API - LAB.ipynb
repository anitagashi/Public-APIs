# API 1

```python
import requests
import pandas as pd
```

```python
response= requests.get('https://openlibrary.org/search.json?q=the+lord+of+the+rings')

book = pd.DataFrame(response.json())
book.head()
#data.shape

```

```python
book.shape
```

```python
book['q'].unique()
```

```python
def flatten(book, col_list):
    for column in col_list:
        flattened = pd.DataFrame(dict(book[column])).transpose()
        columns = [str(col) for col in flattened.columns]
        flattened.columns = [column + '_' + colname for colname in columns]
        book = pd.concat([book, flattened], axis=1)
        book = book.drop(column, axis=1)
    return book
```

```python
dict(book['docs'])
pd.DataFrame(dict(book['docs']))
pd.DataFrame(dict(book['docs'])).transpose()
```

```python
nested_columns = ['docs']

flat = flatten(book,nested_columns)
flat
```
